מערך הנתונים של SLURP

(Spoken Language Understanding Resource Package)

הוא אוסף מקיף שנועד להקל על מחקר ופיתוח במערכות הבנת השפה המדוברת

סקירה כללית של מערך הנתונים של SLURP:

מטרה: SLURP מספקת קבוצה עשירה של הקלטות אודיו בשילוב עם הערות סמנטיות, המאפשרות פיתוח והערכה של מודלים של SLU שיכולים לפרש פקודות ושאילתות מדוברות.

תוכן: מערך הנתונים כולל כ-72,000 הקלטות אודיו של אינטראקציות של משתמשים עם עוזר וירטואלי, בכל אחת מהן מובאות:

תמלול: ייצוגים טקסטואליים של האמירות המדוברות.
כוונות: המטרה או הפעולה הבסיסית שהמשתמש מתכוון לבצע (למשל, הגדרת אזעקה, שאילתת מזג האוויר).
ישויות: פרטים ספציפיים בתוך האמירה, כגון תאריכים, שעות או שמות.
דומיינים: SLURP משתרע על פני 18 דומיינים נפרדים, כולל:

ניהול אזעקות
פניות לגבי מזג האוויר
שליטה במוזיקה
תזמון לוח שנה
אינטראקציות של מכשיר בית חכם
הערות: כל קובץ שמע מסומן בקפידה כדי לכלול:

המשפט הנאמר.
הערות סמנטיות המדגישות כוונות וישויות.
תגיות חלקי דיבור לכל מילה.
מטא נתונים כגון תנאי הקלטה ומידע על דובר.
בקשות פוטנציאליות לסיוע לקשישים:

מינוף SLURP יכול להיות מועיל במיוחד בפיתוח יישומי עוזר קולי המותאמים לקשישים, שכן הוא מציע:

תרחישי אינטראקציה מגוונים: מגוון הדומיינים הרחב של מערך הנתונים משקף משימות נפוצות שמשתמשים מבוגרים עשויים לבצע, כגון הגדרת תזכורות, שליטה במכשירים ביתיים או חיפוש מידע.

שונות שפה טבעית: SLURP לוכדת מגוון דרכים שבהן משתמשים עשויים לנסח את הבקשות שלהם, ומסייעת ביצירת מערכות שמבינות דפוסי דיבור ומינוחים שונים.

הדרכה והערכה: מערך הנתונים משמש בסיס חזק לאימון מודלים של למידת מכונה והערכת הביצועים שלהם בתרחישים בעולם האמיתי.


```python
# Check the size of training and test subsets
train_size = len(ds['train'])
test_size = len(ds['test'])

print(f"Training set size: {train_size} examples")
print(f"Test set size: {test_size} examples")
```

```python
train_df = ds['train'].to_pandas()
display(train_df)
```

```python
print(train_df.columns)
```

```python
# Check for null values in the entire dataframe
null_values = train_df.isnull().sum()

# Print the count of null values per column
print("Null values per column:")
print(null_values)

# Check for NaN values in the entire dataframe
nan_values = train_df.isna().sum()

# Print the count of NaN values per column
print("NaN values per column:")
print(nan_values)

```

```python
import numpy as np

# Remove the 'audio' column
train_df = train_df.drop(columns=['audio'])

# Identify columns with numpy.ndarray objects
ndarray_cols = [col for col in train_df.columns if train_df[col].apply(lambda x: isinstance(x, np.ndarray)).any()]

# Convert numpy.ndarray objects to strings
for col in ndarray_cols:
    train_df[col] = train_df[col].apply(lambda x: str(x) if isinstance(x, np.ndarray) else x)

# Now, drop duplicate rows based on all remaining columns
distinct_records = train_df.drop_duplicates()

# Print the size of the distinct_records (rows, columns)
print(f"Size of distinct_records: {distinct_records.shape[0]} rows and {distinct_records.shape[1]} columns")

# Display the distinct records
display(distinct_records)
```

```python
import matplotlib.pyplot as plt

# Analyze intent distribution
intent_counts = distinct_records['intent'].value_counts()

# Create a larger figure for clarity
plt.figure(figsize=(20, 10))  # Increase figure size

# Plot the full distribution of intents
intent_counts.plot(kind='bar')

# Customize the plot
plt.title('Distribution of Intents in SLURP Dataset', fontsize=16)
plt.xlabel('Intent', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=90, ha='center', fontsize=12)  # Rotate labels by 90 degrees for better readability

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# List of intents relevant for Siri
siri_intents = [
    'calendar_set', 'calendar_query', 'weather_query', 'play_music', 'play_audiobook',
    'play_radio', 'email_query', 'sendemail', 'alarm_query', 'alarm_set',
    'datetime_query', 'social_post', 'qa_factoid', 'general_query', 'news_query'
]

# Filter the dataframe to keep only rows with these intents
siri_filtered_df = distinct_records[distinct_records['intent'].isin(siri_intents)]

# Print the size of the filtered dataframe (rows, columns)
print(f"Filtered table size: {siri_filtered_df.shape[0]} rows and {siri_filtered_df.shape[1]} columns")

# Display the filtered records
display(siri_filtered_df)
```

```python
# Print the size of the filtered dataframe (rows, columns)
print(f"Filtered table size: {siri_filtered_df.shape[0]} rows and {siri_filtered_df.shape[1]} columns")

# Analyze the distribution of the filtered intents
siri_intent_counts = siri_filtered_df['intent'].value_counts()

# Create a DataFrame for the table
siri_intent_table = siri_intent_counts.reset_index()
siri_intent_table.columns = ['Intent', 'Count']

# Display the filtered intent distribution as a table
print("Filtered Intent Distribution for Siri:")
print(siri_intent_table.to_string(index=False))
```

```python
import matplotlib.pyplot as plt

# Analyze intent distribution
intent_counts = siri_filtered_df['intent'].value_counts()

# Create a larger figure for clarity
plt.figure(figsize=(20, 10))  # Increase figure size

# Plot the full distribution of intents
intent_counts.plot(kind='bar')

# Customize the plot
plt.title('Distribution of Intents in SLURP Dataset', fontsize=16)
plt.xlabel('Intent', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=90, ha='center', fontsize=12)  # Rotate labels by 90 degrees for better readability

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create 'word_count' and 'transcript_length' columns using .loc to avoid SettingWithCopyWarning
siri_filtered_df.loc[:, 'word_count'] = siri_filtered_df['transcript'].apply(lambda x: len(x.split()))  # Word count
siri_filtered_df.loc[:, 'transcript_length'] = siri_filtered_df['transcript'].apply(len)  # Transcript length in characters

# Create a figure with 1 row and 2 columns for subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot the distribution of word counts on the first subplot
axes[0].hist(siri_filtered_df['word_count'], bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Word Counts in Transcripts')
axes[0].set_xlabel('Word Count')
axes[0].set_ylabel('Frequency')

# Plot the distribution of transcript lengths on the second subplot
axes[1].hist(siri_filtered_df['transcript_length'], bins=30, color='skyblue', edgecolor='black')
axes[1].set_title('Distribution of Transcript Lengths')
axes[1].set_xlabel('Transcript Length (characters)')
axes[1].set_ylabel('Frequency')

# Adjust the layout for better spacing
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all the transcripts into one large string
text = ' '.join(siri_filtered_df['transcript'])

# Create a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Plotting transcript length by intent
plt.figure(figsize=(12, 6))
siri_filtered_df.groupby('intent')['transcript_length'].mean().sort_values().plot(kind='barh', color='skyblue')
plt.title('Average Transcript Length by Intent')
plt.xlabel('Average Transcript Length')
plt.ylabel('Intent')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each intent
intent_counts = siri_filtered_df['intent'].value_counts()

# Plot the distribution of intents
plt.figure(figsize=(12, 6))
sns.barplot(x=intent_counts.index, y=intent_counts.values, palette='Blues_d')
plt.title('Class Distribution of Intents')
plt.xlabel('Intent')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Check for underrepresented intents (if any intent has very low counts)
threshold = 50  # You can adjust the threshold based on your data
underrepresented_intents = intent_counts[intent_counts < threshold]

# Print underrepresented intents
print("Underrepresented intents (with fewer than", threshold, "occurrences):")
print(underrepresented_intents)
```
